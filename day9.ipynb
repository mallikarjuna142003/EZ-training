```python
class node:
    def __init__(self,data):
        self.value=data
        self.left=None
        self.right=None
        
def preorder(root):
    if root == None:
        return

    print(root.value)
    preorder(root.left)
    preorder(root.right)
    
def postorder(root):
    if root == None:
        return

    print(root.value)
    preorder(root.right)
    preorder(root.left)    
        
if __name__ == "__main__":
    root = node(1)
    
    root.left = node(2)
    root.right = node(3)
    
    root.left.left=node(4)
    root.left.right=node(5)
    
    root.right.left=node(6)
    root.right.right=node(7)
    print("preorder\n")
    preorder(root)
    print("\npostorder")
    postorder(root)
```

```python
class node_data:
    def __init__(self, Node, Hkey):
        self.node = Node
        self.hkey = Hkey
        
def topview(root):
    if not root:
        return

    temp = node_data(root, 0)
    Q = [temp]
    
    key_Dict = {}
    
    while Q:
        curr = Q.pop(0)
        
        if curr is None:
            if not Q:
                break
            else:
                print()
                Q.append(None)
        else:
            if curr.hkey not in key_Dict:
                key_Dict[curr.hkey] = curr.node.value
                
            if curr.node.left is not None:
                temp = node_data(curr.node.left, curr.hkey - 1)
                Q.append(temp)
                
            if curr.node.right is not None:
                temp = node_data(curr.node.right, curr.hkey + 1)
                Q.append(temp)
    
    for i in sorted(key_Dict):
        print(key_Dict[i])
            
    print(key_Dict)

class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

if __name__ == "__main__":
    root = Node(1)
    root.left = Node(2)
    root.right = Node(3)
    root.left.left = Node(4)
    root.left.right = Node(5)
    root.right.left = Node(6)
    root.right.right = Node(8)
    root.left.right.left = Node(9)
    root.left.right.right = Node(10)
    root.right.right.right = Node(11)
    root.left.right.left.left = Node(12)
    root.left.right.left.right = Node(13)

    topview(root)
```

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

class NodeData:
    def __init__(self, node, hkey):
        self.node = node
        self.hkey = hkey

def sideview(root):
    if not root:
        return
    
    Q = [NodeData(root, 0)]
    key_Dict = {}
    
    while Q:
        curr = Q.pop(0)
        
        # For side view, we keep the last value encountered at each horizontal level
        key_Dict[curr.hkey] = curr.node.value
                
        if curr.node.left is not None:
            Q.append(NodeData(curr.node.left, curr.hkey - 1))
                
        if curr.node.right is not None:
            Q.append(NodeData(curr.node.right, curr.hkey + 1))
    
    # Sorting the dictionary keys and printing the corresponding values
    for hkey in sorted(key_Dict.keys()):
        print(key_Dict[hkey])
            
    # Optional: Print the key dictionary if needed
    print(key_Dict)

if __name__ == "__main__":
    root = Node(1)
    root.left = Node(2)
    root.right = Node(3)
    root.left.left = Node(4)
    root.left.right = Node(5)
    root.right.left = Node(6)
    root.right.right = Node(8)
    root.left.right.left = Node(9)
    root.left.right.right = Node(10)
    root.right.right.right = Node(11)
    root.left.right.left.left = Node(12)
    root.left.right.left.right = Node(13)

    sideview(root)
```
